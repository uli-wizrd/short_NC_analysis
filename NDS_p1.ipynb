# Procesamiento y visualización de la información 

```python
# 1 Transformamos la constante en una lista de sus digitos

d_1, *r_d = str(c_n).split('.')
c_1,*u_r_d = r_d[0]
u_r_d.insert(0,c_1)
l_m = list(map(int,u_r_d))
l_m.insert(0,int(d_1))

#print('Lista de los primeros {} digitos de Pi: \n {}'.format(re_d+1, l_m))
```

```python
# 2 Ahora crearemos la funcion que cuenta cuantos de cada digito hay

def d_count(dl):
    try:
        if type(dl) != list:
            raise Exception('not a list')
    except Exception as inst_1:
        print('Error: input type is {}, please select a list as input next time!'.format(inst_1))
    else:
        r = []
        for i in range(10):
            if i in dl:
                oc = dl.count(i)
                l_p = [i,oc]
                r.append(l_p)
            else:
                pass
        return r
```

```python
# 3 Aqui definimos lo que necesitamos para imprimir cosas 
# Introducimos un poco de magia negra

bd = d_count(l_m)
bd =sorted(bd,key=lambda bd: bd[1])

di = [str(x[0]) for x in bd]
print('Etiquetas para graficas: \n {}'.format(di))
ocr = [x[1] for x in bd]
print('Lista de veces que aparece un digito: \n {}'.format(ocr))

m_ocr = min(ocr)
l_d = []

for i in ocr:
    d_1 = i - m_ocr
    l_d.append(d_1)

print('Delta de frecuencia en relacion al valor minimo en la lista: \n {}'.format(l_d))
```

```python
# 4 creamos funciones para colores y porcentajes

# Debemos ajustarla en el rango de 0,1

def porc(l,aes=True):

    try:
        if type(l) != list:
            raise Exception('not a list')
    except Exception as inst_1:
        print('Error! the input is {}, please use the correct data type'.format(inst_1)) 
    
    else:
        if aes:
            m = max(l)
        else:
            m = sum(l)
        nl = []

        for i in l:
            p = round(i/m, 4)
            nl.append(p)

        return nl

def mp_clr(l, c=0.75, m='viridis'):

    cmap = matplotlib.cm.get_cmap(m)

    vs = porc(l)

    l_1 = list(map(cmap, vs))
    l_2 = np.array(l_1)
    l_2[:, 0:3] *= c

    return l_2
```

```python
# 5 Definimos el arreglo que tendra nuestros colores
arr_mc = mp_clr(l_d)
```

```python
# 8 Media aritmética
m_ar = lambda l: round(sum(l)/len(l),4)
m_a = round(m_ar(ocr),4)
print('Valor promedio de aparición de digitos: \n{}'.format(m_a))
```

```python
# 9 Función para obtener la desviación estandar de una población

def s_dv(m_a,l_v):
    sqs = []
    for i in l_v:
        sq = round((m_a - i)**2, 4)
        sqs.append(sq)
    s_sq = sum(sqs)
    Nu = len(sqs)
    return round(math.sqrt(s_sq/Nu),4)
```

# Justificación

La motivacion detras de todo esto surgió al analizar la siguiente secuencia númerica de los dígitos del señor Pi:<br>
3.141592653589793238462643383279502884197169399375105820974944592307816406286208998628034825342117067979 <br>
Podemos encontrar el siguiente patrón: xyx (donde x así como y representan un entero positivo) en la constante <br>
Como podemos ver a continuación:<br>
3.[141] 5926 [535] 8 [979] [323] 84 [626] 43 [383] 27950288419716 [939] 937510582097 [494]45923078.<br>
En la siguiente sección se hace una exploración de este patrón y sus componentes.

```python
# Algoritmo y lista de la distancia entre cada hueco

g_sl = []
l_sim = []
s_v = 1
g_c = 0
l_r = len(l_m)
ct = 0

while l_r != 0:

    fp = ct+1+s_v
    #print('Currval:' + str(l_m[ct]))
    #print(g_c)
    if fp >= len(l_m):
        break

    if l_m[ct] == l_m[fp]:

        #Nos llevamos la simetria
        xyx = [l_m[ct],l_m[fp-1],l_m[fp]]
        l_sim.append(xyx)

        g_sl.append(g_c)
        g_c = 0
        nv = 3
        #print('seq begin:' +str(l_m[ct]))
        ct = ct + nv
        l_r = l_r - nv
               #print("ct:" + str(ct))
        #print("l_r:" + str(l_r))
    else:
        g_c += 1
        ct += 1
        l_r = l_r - 1 
```

```python
# Procesamiento de lista de patrones

arr_sim = np.array(l_sim)

xvs = list(arr_sim[:,0:1].flatten())
yvs = list(arr_sim[:,1:2].flatten())

xs = np.array(sorted(d_count(xvs), key=lambda l: l[1]))
print('Dígitos X en patrones XYX y su frecuencia: \n {} \n {}'.format(xs[:,0],xs[:,1]))
ys = np.array(sorted(d_count(yvs), key=lambda l: l[1]))
print('\n Dígitos Y en patrones XYX y su frecuencia: \n {} \n {}'.format(ys[:,0],ys[:,1]))
```

# Análisis de patrones XYX

A continuación se enumeran los pasos que se tomaron para procesar la información de los patrones XYX en la secuencia númerica analizada.

```python
# 1 convertimos en texto nuestros patrones
ts_l = []
top_pat = []
for i in l_sim:
    dl_1 = []
    for j in range(len(i)):
        tdv = str(i[j])
        dl_1.append(tdv)
    ts_l.append(int(''.join(dl_1)))
```

```python
# 2 Cuantas veces aparece este patrón
u_vs = []
tpat_l = []

for i in ts_l:
    for k in range(len(ts_l)):
        if ts_l[k] not in u_vs:
            u_vs.append(ts_l[k])
        else:
            pass
print('Cantidad de patrones XYX únicos: \n {}'.format(len(u_vs)))

for i in u_vs:
    tms = ts_l.count(i)
    tpat_l.append([i,tms])

stpat_l = sorted(tpat_l,key=lambda l: l[1],reverse=True)
stpat_l = stpat_l[0:4]
print('Lista de los 5 patrones más frecuentes: \n {}'.format(stpat_l))
stpat_l = np.array(stpat_l)

tms_g = stpat_l[:,1]
patt = list(map(str,stpat_l[:,0]))
```
